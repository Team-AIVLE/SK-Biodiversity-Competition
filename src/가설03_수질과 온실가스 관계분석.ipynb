### **수질 데이터 (하천/호소)**

호소의 수질 데이터

```python
hoso = pd.read_csv(pjoin(PROC_DATA, "호소.csv"), parse_dates=['date'])
hoso
```

하천의 수질 데이터

```python
river = pd.read_csv(pjoin(PROC_DATA, "하천.csv"), parse_dates=['date'])
river
```

```python
river['name'].unique()
```

수원시 전체 하천, 호소의 수질 데이터

```python
water = pd.concat([hoso, river], ignore_index=True)
water['month'] = water['date'].dt.month
water['year'] = water['date'].dt.year

water
```

온실가스 데이터

```python
air = pd.read_csv(pjoin(PROC_DATA, "온실가스.csv"), encoding="cp949")
air
```

```python
air['month'] = air['일시'].apply(lambda x: int(x.split("-")[-1]))
air['year'] = air['일시'].apply(lambda x: int(x.split("-")[0]))
```

연도 및 월에 따라 데이터 병합

```python
data = pd.merge(water, air, on = ['month', 'year'])
data
```

결측치 확인

```python
data.isna().sum()
```

데이터 형변환: NH3-N 데이터를 수치형으로 변환

```python
data.info()
```

```python
def convert_to_float(x):
    try:
        return float(x)
    except:
        return np.nan
```

```python
data['NH3-N (MG/L)'] = data['NH3-N (MG/L)'].apply(convert_to_float)
```

```python
import scipy.stats as spst

cols = ['PH', 'SS', 'T-N', 'T-P', 'CHL-A', 'COD', 'TOC', 'DO', 'EC (MS/CM)', 'NH3-N (MG/L)', '수온', 'BOD']
    
def get_correlation_with_water_quality(data, tgt):
    result = pd.DataFrame()
    for c in cols:
        try:
            sub_data = data[[tgt, c]].dropna()
            if len(sub_data) < 5: continue

            coe, pval = spst.pearsonr(sub_data[c], sub_data[tgt])
            result = result.append({"수질 지표" : c,
                                    "상관계수" : coe,
                                   "p-value" : pval}, ignore_index=True)
        except Exception as e:
            print(e)
    return result

def show_scatter_graph_with_water_quality(data, tgt):
    plt.figure(figsize=(20, 10))
    for i in range(12):
        plt.subplot(3, 4, i + 1)
        sns.regplot(x=tgt, y=cols[i], data=data)
    plt.show()
```

## **평균 CO2 배경 대기 농도와 수질의 상관관계**

```python
show_scatter_graph_with_water_quality(data, tgt='평균 이산화탄소(CO2) 배경대기농도(ppm)')
```

```python
result = get_correlation_with_water_quality(data, tgt='평균 이산화탄소(CO2) 배경대기농도(ppm)')
result[result['p-value'] < 0.05].sort_values(by=['상관계수'])
```
